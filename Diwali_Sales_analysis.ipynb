# Data Cleaning

```python
# import necessary library
# numercial computation
import numpy as np

# data manipulation
import pandas as pd

# for visualization
import matplotlib.pyplot as plt

# for advance visualization
import seaborn as sns

# for avoid warning
import warnings
warnings.filterwarnings("ignore")
```

```python
# load data
df=pd.read_csv("Diwali Sales Data.csv",encoding='unicode_escape')
# chechk top records from dataset
df.head()
```

```python
# check bottom data from dataset
df.tail()
```

```python
# chechk shape of dataset (rows/columns)
df.shape
```

```python
# chechk data type,columns,rows,null or not null value
df.info()
```

```python
# chechk columns in dataset
df.columns
```

```python
# delete unrelated columns
df.drop(["Status",'unnamed1'],axis=1,inplace=True)
# axis=1 means delete columns and axis=0 means delete rows
# inplace=True means saved in dataset
```

```python
df.head()
```

The unessary column are deleted sucessfully

```python
# Check null values
df.isnull().sum()
```

```python
df.shape
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
# check shape
df.shape
```

```python
df.isnull().sum()
```

The null values delete sucessfully

```python
# change data type of amount as int
df['Amount']=df['Amount'].astype('int')
```

```python
df.info()
```

```python
df['Amount'].dtypes
```

```python
df.describe()
```

```python
# describe specific data
df[["Age","Orders","Amount"]].describe()
```

# ERD

## Gender

```python
sns.countplot(x=df['Gender'],data=df)
plt.show()
```

```python
sns.barplot(data=df, x='Gender', y='Amount',palette="deep")
plt.title("Gender vs Amount")
plt.ylabel("Amount")
plt.xlabel("Gender")
plt.show()
```

```python
df['Gender'].value_counts()
```

## Zone

```python
df['Zone'].value_counts()
```

```python
sns.barplot(data=df, x='Zone', y='Amount',estimator=sum,palette="Set2")
plt.title("Zone vs Amount")
plt.ylabel("Amount")
plt.xlabel("Zone")
plt.xticks(rotation=90)  # or try 60 or 90 degrees
plt.show()
```

```python
df['Age Group'].value_counts()
```

```python
sns.barplot(data=df, x='Age Group', y='Amount',estimator=sum,order=sorted(df['Age Group'].unique()),palette="muted")
plt.title("Age Group vs Total Purchase Amount")
plt.ylabel("Total Amount")
plt.xlabel("Age Group")
plt.show()
```

```python
df['Occupation'].value_counts()
```

```python
sns.countplot(x=df['Occupation'],data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
sns.barplot(data=df, x='Occupation', y='Amount',estimator=sum, palette="pastel")
plt.title("Occupation vs Amount")
plt.ylabel("Amount")
plt.xlabel("Occupation")
plt.xticks(rotation=90)
plt.show()
```

```python
df['Marital_Status'].value_counts()
```

```python
sns.barplot(data=df, x='Marital_Status', y='Amount', estimator=sum,palette="deep")
plt.title("Marital_Status vs Total Purchase Amount")
plt.ylabel("Total Amount")
plt.xlabel("Marital_Status")
plt.show()
```

```python
df['State'].value_counts()
```

```python
sns.barplot(data=df, x='State', y='Amount', estimator=sum,palette="dark")
plt.title("Zone vs Total Purchase Amount")
plt.ylabel("Total Amount")
plt.xlabel("State")
plt.xticks(rotation=90)  # or try 60 or 90 degrees
plt.show()
```

```python
df['Product_Category'].value_counts()
```

```python
sns.barplot(data=df, x='Product_Category', y='Amount',estimator=len,palette="colorblind")
plt.title("Zone vs Total Purchase Amount")
plt.ylabel("Total Amount")
plt.xlabel("Zone")
plt.xticks(rotation=90)  # or try 60 or 90 degrees
plt.show()
```
